```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([1.1, 5.0, 2.8, 7.3])  # evidence for each choice
theta = 2.0                         # determinism parameter

ps = np.exp(X * theta)
ps /= np.sum(ps)
```

```python
plt.plot(X, ps)
```

```python
X = np.array([
    [1.1, 5.0, 2.2, 7.3],
    [6.5, 3.2, 8.8, 5.3],
    [2.7, 5.1, 9.6, 7.4],
    ])  

# looping through rows of X
ps1 = np.empty(X.shape)
for i in range(X.shape[0]):
    ps1[i,:]  = np.exp(X[i,:] * theta)
    ps1[i,:] /= np.sum(ps1[i,:])
```
